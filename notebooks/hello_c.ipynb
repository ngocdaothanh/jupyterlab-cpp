# Learning C language in 2025

## Why should I learn C language?

- For learning the first language in the category "system language"; working close to the hardware.
- C is the simplest and easiest language in the category "system language".
- C is the standard FFI; we need to learn a bit of C language even when we use other system languages like C++, Rust, Golang.

## Steps to create, compile, and run a C program

A C program always have a `main` function. It's the entrypoint for the operating system to run the program.

Basic C program `hello_c.c`:

```C
// For having printf function below
#include <stdio.h>

// Signature of the main function
int main() {
    printf("Hello");
    return 0;
}
```

To compile the C language source code file `hello_c.c` to executable binary program `a.out` for the operating system to run:

```sh
gcc hello_c.c
```

gcc is a very popular open source compiler.

To run the executable binary program `a.out` in the current directory:

```sh
./a.out
```
